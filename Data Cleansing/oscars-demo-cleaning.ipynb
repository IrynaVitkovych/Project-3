```python
#dependencies 
import pandas as pd
import pandera as pa
```

```python
# pull in oscars demographic data into dataframe
oscars_demo_df = pd.read_csv('../Raw Data/Oscars-demographics-DFE.csv', encoding= 'unicode_escape')

oscars_demo_df.head(10)
```

```python
# check for null values
oscars_demo_df.info()
```

```python
# since many columns contain nearly all nulls, we will drop the irrelevant ones
oscars_demo_df= oscars_demo_df.drop(columns=['_unit_id','_golden','_unit_state','_trusted_judgments','_last_judgment_at','birthplace:confidence','date_of_birth:confidence','race_ethnicity:confidence','religion:confidence','sexual_orientation:confidence','year_of_award:confidence','birthplace_gold','date_of_birth_gold','race_ethnicity_gold','religion_gold','sexual_orientation_gold','year_of_award_gold'])


```

```python
#check to for null values once columns have been removed
oscars_demo_df.info()
```

```python
# Define schema for dataframe using Pandera
schema = pa.DataFrameSchema({
    "birthplace": pa.Column(pa.String, nullable=False),
    "date_of_birth": pa.Column(pa.String, nullable=False),
    "race_ethnicity": pa.Column(pa.String, nullable=False),
    "religion": pa.Column(pa.String, nullable=False),
    "sexual_orientation": pa.Column(pa.String, nullable=False),
    "year_of_award": pa.Column(pa.Int, nullable=False),
    "award": pa.Column(pa.String, nullable=False),
    "biourl": pa.Column(pa.Object, nullable=False),
    "movie": pa.Column(pa.String, nullable=False),
    "person": pa.Column(pa.String, nullable=False),
})

#print the schema
print(schema)
```

```python
#check dataframe for errors
schema.validate(oscars_demo_df)
```

```python
# update name of column to align with column name of another dataframe we will utilize for our tables
oscars_demo_df.rename(columns={'person':'name'}, inplace=True)

oscars_demo_df.head()
```

```python
#update column name to match others in dataframe, again
oscars_demo_df.rename(columns={'name':'winner'}, inplace=True)

oscars_demo_df
```

```python
# Save file to import into Postgres as csv file
oscars_demo_df.to_csv("../Clean Data CSV/oscar_winner_demographics.csv")
```
