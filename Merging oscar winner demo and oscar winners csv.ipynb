```python
import pandas as pd
```

```python
owd = pd.read_csv("Clean Data/oscar_winner_demographics.csv")
owd.rename(columns={'birthplace': 'Birthplace'}, inplace=True)
owd.rename(columns={'date_of_birth': 'Date_Of_Birth'}, inplace=True)
owd.rename(columns={'race_ethnicity': 'Race_Ethnicity'}, inplace=True)
owd.rename(columns={'religion': 'Religion'}, inplace=True)
owd.rename(columns={'sexual_orientation': 'Sexual_Orientation'}, inplace=True)
owd.rename(columns={'year_of_award': 'Year_Of_Award'}, inplace=True)
owd.rename(columns={'award': 'Award'}, inplace=True)
owd.rename(columns={'biourl': 'Biourl'}, inplace=True)
owd.rename(columns={'Movie Name': 'Movie_Name'}, inplace=True)
owd.rename(columns={'winner': 'Winner'}, inplace=True)
owd.head(10)
```

```python
ow = pd.read_csv("Clean Data/oscar_winners.csv")
ow.rename(columns={'Year Film': 'Year_Film'}, inplace=True)
ow.rename(columns={'Year Ceremony': 'Year_Ceremony'}, inplace=True)
ow.rename(columns={'Name of Winner': 'Winner'}, inplace=True)
ow.rename(columns={'Movie Name': 'Movie_Name'}, inplace=True)
ow.rename(columns={'Oscar Award ID': 'Oscar_Award_ID'}, inplace=True)
ow.head(10)
```

```python
# Merge the above two dataframes
merged_df = pd.merge(owd,ow, on=['Winner', 'Movie_Name'])

# Drop extra columns from the ow dataframe
merged_df = merged_df.drop(columns=['Year_Film','Year_Ceremony','Award_y','Oscar_Award_ID'])

# Rename column to correct name
merged_df.rename(columns={'Award_x': 'Award'}, inplace=True)

# Add artificial ID for primary key
merged_df['Oscar_Winner_Demo_ID'] = range(1, len(merged_df) + 1)

# Reposition column to beginning of dataframe
merged_df.insert(0, 'Oscar_Winner_Demo_ID', merged_df.pop('Oscar_Winner_Demo_ID'))

merged_df.head(20)
```

```python
# Save to Clean Data folder
merged_df.to_csv("Clean Data/Oscar_Winner_Combined.csv")
```
