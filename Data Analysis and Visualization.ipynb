```python
import pandas as pd
import numpy as np
```

```python
# Creata Data to compare Oscar & Golden Globe Winners

oscar_winners = pd.read_csv('CleanData/Oscar_Winners.csv', encoding='utf-8')
goldenglobe_winners = pd.read_csv('CleanData/Golden_Globe_Winners.csv', encoding='utf-8')


# Merge the datasets on specified columns
awards_data_df = pd.merge(oscar_winners, goldenglobe_winners, how='outer', on=["Movie Name"] )

# Display the merged DataFrame
awards_data_df.head(5)
```

```python
# Drop unwanted columns 
columns_to_drop = ['Oscar Winner ID', 'Oscar Award ID', 'Year Film', 'Golden Globe Winner ID', 'GG Award ID']
oscar_goldenglobe_df = awards_data_df.drop(columns=columns_to_drop)

# Rename columns
oscar_goldenglobe_df.rename(columns={'Winner_x': 'Oscar Winner', 'Award': 'Oscar Award', 'Year Ceremony_x': 'Year Ceremony', 'Winner_y': 'GG Winer', 'Year Ceremony_y': 'Year'}, inplace=True)

# Display the updated DataFrame
oscar_goldenglobe_df.head(5)
```

```python
# Remove the index column
df_without_index = oscar_goldenglobe_df.reset_index(drop=True)

# Reorder the columns 
columns_order = ['Movie Name', 'Year Ceremony', 'Oscar Award', 'Oscar Winner', 'Year', 'Golden Globe Award', 'GG Winer']
new_oscar_goldenglobe_df = df_without_index[columns_order]

# Convert columns to integers
columns_to_convert = ['Year Ceremony', 'Year',]
new_oscar_goldenglobe_df [columns_to_convert] = new_oscar_goldenglobe_df [columns_to_convert].fillna(0).astype(int)
new_oscar_goldenglobe_df.head(5)
```

```python
print(len(new_oscar_goldenglobe_df))
```

```python
unique_movies = new_oscar_goldenglobe_df['Movie Name'].unique()
print(unique_movies)
print(len(unique_movies))
```

```python
best_picture_movies = new_oscar_goldenglobe_df[
    (new_oscar_goldenglobe_df['Oscar Award'] == 'BEST PICTURE') &
    ((new_oscar_goldenglobe_df['Golden Globe Award'] == 'Best Motion Picture - Drama') |
    (new_oscar_goldenglobe_df['Golden Globe Award'] == 'Best Motion Picture - Musical or Comedy'))
]
best_picture_movies.head(20)
```

```python
# Export list of Oscar Winners as CSV files.
new_oscar_goldenglobe_df.to_csv("CleanData/Only_Oscar_GG_Award_Movies.csv", index=False)
```
