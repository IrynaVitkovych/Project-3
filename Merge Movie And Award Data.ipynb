```python
import pandas as pd
import numpy as np
```

```python
# Files to use
movies_data = pd.read_csv('Resources/movies.csv', encoding='utf-8')
oscar_winners = pd.read_csv('CleanData/Oscar_Winners.csv', encoding='utf-8')
goldenglobe_winners = pd.read_csv('CleanData/GoldenGlobe_Winners.csv', encoding='utf-8')

# Combine the data into a single DataFrame
combine_data_df = pd.merge(movies_data, oscar_winners, how='outer', on=["Movie_Name", "Movie_Name"])
complete_data_df = pd.merge(combine_data_df, goldenglobe_winners, how='outer', on=["Movie_Name", "Movie_Name"])

complete_data_df.head(10)
```

```python
# Columns name info
complete_data_df.columns
```

```python
# Drop unwanted columns 
columns_to_drop = ['Unnamed: 0', 'Director', 'Oscar_Winner_ID', 'Winner_x', 'Award', 'Year_Film', 'Year_Ceremony_x', 'Golden_Globe_Winner_ID', 'Winner_y', 'Year_Ceremony_y','Golden_Globe_Award']
complete_movies_df = complete_data_df.drop(columns=columns_to_drop)

# Identify which columns contain non-integer  values
columns_to_convert = ['Votes', 'Meta_score', 'Year', 'Movie_ID', 'Director_ID']

# Clean the columns
for col in columns_to_convert:
    #Replace NaN values with zeros
    complete_movies_df[col] = complete_movies_df[col].fillna(0)  # Replace NaN with 0
    complete_movies_df[col] = complete_movies_df[col].astype(float)  # Convert to float to handle potential NaN or inf values

# Convert the columns to integers
complete_movies_df[columns_to_convert] = complete_movies_df[columns_to_convert].astype(int)

complete_movies_df.head(10)
```

```python
# Drop duplicate rows
movies_df = complete_movies_df.drop_duplicates(subset=['Movie_Name', 'Oscar_Award_ID', 'GG_Award_ID'])
movies_df.head(10)
```

```python
unique_movies = movies_df['Movie_Name'].unique()
print(unique_movies)
print(len(unique_movies))
```

```python
# Export list of Oscar Winners as CSV files.
movies_df.to_csv("CleanData/Movies.csv", index=False)
```

#### Now, when we have movies ids for all movies, we can update our Movies Dataset.

```python
## Files to use
movies_id = pd.read_csv('Resources/All_Movies_IDs.csv', encoding='utf-8')
movies_data = pd.read_csv('Resources/movies.csv', encoding='utf-8')

# Combine the data into a single DataFrame
nmd_data_df = pd.merge(movies_id, movies_data, how='outer', on=["Movie_Name", "Movie_Name"])


nmd_data_df.head(5)
```

```python
# Drop unwanted columns 
columns_to_drop = ['Unnamed: 0', 'Movie_ID_y']
nmd_data_df = nmd_data_df.drop(columns=columns_to_drop)

# Identify which columns contain non-integer  values
columns_to_convert = ['Votes', 'Meta_score', 'Year', 'Director_ID']

# Clean the columns
for col in columns_to_convert:
    #Replace NaN values with zeros
    nmd_data_df[col] = nmd_data_df[col].fillna(0)  # Replace NaN with 0
    nmd_data_df[col] = nmd_data_df[col].astype(float)  # Convert to float to handle potential NaN or inf values

# Convert the columns to integers
nmd_data_df[columns_to_convert] = nmd_data_df[columns_to_convert].astype(int)

# Rename the column "Rearranged Movie Name" to "Movie Name"
nmd_data_df.rename(columns={"Movie_ID_x": "Movie_ID"}, inplace=True)

nmd_data_df.head(10)
```

```python
# Export list of Oscar Winners as CSV files.
nmd_data_df.to_csv("Resources/Complete_Movies_Table.csv", index=False)
```
