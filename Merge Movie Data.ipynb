```python
import pandas as pd
```

```python
# Files to use
movies_data = pd.read_csv('CleanData/Movies.csv', encoding='utf-8')
oscar_winners = pd.read_csv('CleanData/oscar_winners.csv', encoding='utf-8')
goldenglobe_winners = pd.read_csv('CleanData/goldenglobe_winners.csv', encoding='utf-8')

# Combine the data into a single DataFrame
combine_data_df = pd.merge(movies_data, oscar_winners, how="left", on=["Movie Name", "Movie Name"])
complete_data_df = pd.merge(combine_data_df, goldenglobe_winners, how="left", on=["Movie Name"])
complete_data_df
```

```python
# Columns name info
complete_data_df.columns
```

```python
# Drop unwanted columns 
columns_to_drop = ['Unnamed: 0', 'Director', 'Year Film', 'Year Ceremony', 'Award','Name of Winner', 'Year_y', 'Status', 'Golden Globe Award',]
complete_movies_df = complete_data_df.drop(columns=columns_to_drop)

# Rename column Year_x
complete_movies_df.rename(columns={'Year_x': 'Year'}, inplace=True)

complete_movies_df.head(10)
```

```python
# Replace NaN values with "N/A"
movies_df = complete_movies_df.fillna("N/A")
movies_df.head(25)
```

```python
# Export list of Oscar Winners as CSV files.
movies_df.to_csv("CleanData/MoviesWithAwards.csv", index=False)
```
