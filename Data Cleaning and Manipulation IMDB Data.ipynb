```python
# Dependencies
import pandas as pd
import pandera as pa
```

```python
# Pull in IMDB Movie Data into a dataframe
imdb_movie_data = pd.read_csv('Raw Data/imdb_movie_data_2023.csv')

imdb_movie_data.head(10)
```

```python
# Define the schema
schema = pa.DataFrameSchema({
    "Movie Name": pa.Column(pa.String, nullable=False),
    "Rating": pa.Column(pa.Float, nullable=False),
    "Votes": pa.Column(pa.Float, nullable=False),
    "Meta Score": pa.Column(pa.Float, nullable=False),
    "Genre": pa.Column(pa.String, nullable=False),
    "PG Rating": pa.Column(pa.String, nullable=False),
    "Year": pa.Column(pa.Int, nullable=False),
    "Duration": pa.Column(pa.String, nullable=False),
    "Cast": pa.Column(pa.String, nullable=False),
    "Director": pa.Column(pa.String, nullable=False),
})

# Print the schema
print(schema)
```

```python
# Check data set for errors
schema.validate(imdb_movie_data)
```

```python
# Correct spelling on Movie Name column 
imdb_movie_data.rename(columns={'Moive Name': 'Movie Name'}, inplace=True)
imdb_movie_data.head()
```

```python
# Check data set for errors
schema.validate(imdb_movie_data)
```

```python
# Check for null values
imdb_movie_data.info()

# Remove rows with null values
imdb_movie_data = imdb_movie_data.dropna()

imdb_movie_data.info()
```

```python
# Check data set for errors
schema.validate(imdb_movie_data)
```

```python
# Remove "Unnamed: 0" column
imdb_movie_data = imdb_movie_data.drop(columns=['Unnamed: 0'])

# Possibly drop the 3 IMDB scoring columns
```

```python
# Check data set for errors
schema.validate(imdb_movie_data)
```

```python
# Reset index
validated_df = imdb_movie_data.reset_index(drop=True)
schema.validate(validated_df)
```

```python
# Creating the Actors dataframe for the SQL database
# Split actor names into actors dataframe
actor_names = validated_df['Cast'].str.split(', ', expand=True).stack().reset_index(drop=True)

# Create a new DataFrame with actor names and actor ID
actors_df = pd.DataFrame(actor_names, columns=['Actor'])

# Remove any Duplicate actor names
actors_df.drop_duplicates(subset='Actor', keep='first', inplace=True)
actors_df.reset_index(drop=True, inplace=True)

# Add an ID for the actors
actors_df['Actor ID'] = range(1, len(actors_df) + 1)

actors_df.head(10)

# Save off file for import to postgres database
actors_df.to_csv("Clean Data/Actors.csv")
```

```python
# Creating the Movies dataframe for the SQL database
movies_df = validated_df.drop(columns=['Cast', 'Director','Genre'])

# Remove any Duplicate actor names
movies_df.drop_duplicates(subset='Movie Name', keep='first', inplace=True)
movies_df.reset_index(drop=True, inplace=True)

# Add an ID for the actors
movies_df['Movie ID'] = range(1, len(movies_df) + 1)

# Save off file for import to postgres database
movies_df.to_csv("Clean Data/Movies.csv")
```

```python
# Creating the Genres dataframe for the SQL database
# Split actor names into actors dataframe
Genre_names = validated_df['Genre'].str.split(', ', expand=True).stack().reset_index(drop=True)

# Create a new DataFrame with genre names and genre ID
Genre_df = pd.DataFrame(Genre_names, columns=['Genre'])

# Remove any duplicate genre names
Genre_df.drop_duplicates(subset='Genre', keep='first', inplace=True)
Genre_df.reset_index(drop=True, inplace=True)

# Add an ID for the genres
Genre_df['Genre ID'] = range(1, len(Genre_df) + 1)

# Save off file for import to postgres database
Genre_df.to_csv("Clean Data/Genres.csv")
```

```python
# Creating the Actors dataframe for the SQL database
# Split actor names into actors dataframe
director_names = validated_df['Director'].str.split(', ', expand=True).stack().reset_index(drop=True)

# Create a new DataFrame with actor names and actor ID
directors_df = pd.DataFrame(director_names, columns=['Director'])

# Remove any Duplicate actor names
directors_df.drop_duplicates(subset='Director', keep='first', inplace=True)
directors_df.reset_index(drop=True, inplace=True)

# Add an ID for the actors
directors_df['Director ID'] = range(1, len(directors_df) + 1)

# Add my own row due to data inconsistencies
new_row1 = {'Director': 'Robert Marianetti'}
new_row2 = {'Director': 'Robert Smigel'}
new_row3 = {'Director': 'David Wachtenheim'}
new_row4 = {'Director': 'Yarrow Cheney'}
new_row5 = {'Director': 'Scott Mosier'}
new_row6 = {'Director': 'Aaron Horvath'}
new_row7 = {'Director': 'Michael Jelenic'}
new_row8 = {'Director': 'Pierre Leduc'}
new_row9 = {'Director': 'Chris Buck'}
new_row10 = {'Director': 'Fawn Veerasunthorn'}
new_row11 = {'Director': 'Walt Dohrn'}
new_row12 = {'Director': 'Tim Heitz'}
new_row13 = {'Director': 'Joaquim Dos Santos'}
new_row14 = {'Director': 'Kemp Powers'}
new_row15 = {'Director': 'Justin K. Thompson'}
new_row16 = {'Director': 'Fawn Veerasunthorn'}
new_row17 = {'Director': 'Fawn Veerasunthorn'}

# Save off file for import to postgres database
directors_df.to_csv("Clean Data/Directors.csv")
```
