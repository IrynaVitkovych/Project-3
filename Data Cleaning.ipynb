```python
import pandas as pd
import numpy as np
```

```python
# Pull in IMDB Movie Data into a dataframe
imdb_movie_data = pd.read_csv('Resources/imdb_movie_data_2023.csv')
imdb_movie_data.head(10)                            
```

```python
imdb_movie_data.info()
```

```python
# Remove "Unnamed: 0" column
imdb_movie_data = imdb_movie_data.drop(columns=['Unnamed: 0'])
```

```python
# Check for null values
# Remove rows with null values
imdb_movie_data = imdb_movie_data.dropna()
imdb_movie_data.info()
```

```python
# Rename column to Movie Name
imdb_movie_data.rename(columns={'Moive Name': 'movie_name'}, inplace=True)
imdb_movie_data.head(10)
```

```python
movies = imdb_movie_data['movie_name'].unique()
print(movies)
```

```python
# Create numpy arrays from 1-1753 for the movies
movie_ids = np.arange(1, len(movies) + 1)

print(movie_ids)
```

```python
# Use a list comprehension to add unique "id" to each movie.

m_ids = ['id' + str(m_id) for m_id in range(1, len(movie_ids) + 1)]
print(m_ids)
```

```python
# Create a category DataFrame with the movie_id array as the unique id and movies list as the movie name.
movie_df = pd.DataFrame({
    'movie_id': m_ids,
    'movie_name': movies
})
```

```python
movie_df
```

```python
# Export movie_df as CSV files.
movie_df.to_csv("Resources/movies.csv", index=False)
```
