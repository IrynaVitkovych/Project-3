```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Pull in IMDB Movie Data into a dataframe
goldenglobes = pd.read_csv('Data/goldenglobes (1).csv')

goldenglobes.head(10)
```

```python
goldenglobes.head(10)
```

```python
goldenglobes.shape
```

```python
goldenglobes.describe()
```

```python
# Check for null values
goldenglobes.info()

# Find rows with null values
null_rows = goldenglobes[goldenglobes.isnull().any(axis=1)]
null_rows.head(10)

# Remove null values as they are not related to movie awards
goldenglobes= goldenglobes.dropna()

goldenglobes.info()
```

```python
# Remove awards that contain "Television" in the string
## We are only looking at the movie golden globe awards
goldenglobes = goldenglobes[~goldenglobes['award'].str.contains("Television", na=False)]

goldenglobes.head(25)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Counting nominations
nominee_count = goldenglobes['status'].value_counts()

# Creating the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=nominee_count.index, y=nominee_count.values, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Movie Title')
plt.ylabel('Number of Nominations')
plt.title('Number of Nominations per Movie')
plt.show()
```

```python
# Counting status
status_counts = goldenglobes['status'].value_counts()

# Creating the pie chart
plt.figure(figsize=(7, 7))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen'])
plt.title('Proportion of Nominees to Winners')
plt.show()
```

```python
# Filter for winners only
winners = goldenglobes[goldenglobes['status'] == 'Winner']

# Count occurrences of each award among winners and get the top 10
top_awards = winners['award'].value_counts().nlargest(10)

# Creating the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=top_awards.values, y=top_awards.index, palette='viridis')
plt.xlabel('Number of Wins')
plt.ylabel('Award Category')
plt.title('Top 10 Award Categories by Number of Wins')
plt.show()
```

```python
# Filter for winners only
winners = goldenglobes[goldenglobes['status'] == 'Winner']

# Count occurrences of each title among winners and get the top 10
top_titles = winners['title'].value_counts().nlargest(10)

# Creating the bar chart
plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=top_titles.values, y=top_titles.index, palette='coolwarm')
plt.xlabel('Number of Wins')
plt.ylabel('Title')
plt.title('Top 10 Titles by Number of Wins')

# Adding labels to each bar
for index, value in enumerate(top_titles):
    plt.text(value, index, str(value))  # position text horizontally slightly ahead of the bar end.

plt.show()
```

```python
# Filter for winners only
winners = goldenglobes[goldenglobes['status'] == 'Winner']

# Group by year and count the number of winners per year
winners_by_year = winners['year'].value_counts().sort_index()

# Filtering to the last 10 years if the dataset covers a longer period
last_10_years = winners_by_year.tail(10)

# Creating the line chart
plt.figure(figsize=(12, 6))
lineplot = sns.lineplot(x=last_10_years.index, y=last_10_years.values, marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('Number of Winners')
plt.title('Number of Winners Over the Last 10 Years')

# Adding labels to each point
for year, count in last_10_years.items():
    plt.text(year, count, str(count), color='black', ha='center', va='bottom')

plt.show()
```
