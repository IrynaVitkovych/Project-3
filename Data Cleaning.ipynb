### Golden Globe Award DataFrame
----
**Clean and Manipulate a Golden Globe Data:**
- check for null values
- remane columns
- remove awards that contain "Television"
- remove rows for Nominees and leave only Winners
- drop duplicate rows
- assign IDs for each unique award
- add Ids for each winner
- export the DataFrame as a 'Golden_Globe_Winners.csv' CSV file
- create a Golden Globe Awards DataFrame with the award_id array as the GG_award_id and awards list as the award category.
- reorder the columns 
- create new Data frame for list of Golden Globe Awards with their unique ID's.
- export the DataFrame as a `Golden_Globe_Winners.csv` CSV file.

```python
# Pull in Golden Globe Award Data into a dataframe
goldenglobe_data = pd.read_csv('Resources/goldenglobes.csv', encoding='utf-8')
goldenglobe_data.head(5)
```

```python
# Rename column Movie Name
goldenglobe_data.rename(columns={'title': 'Winner', 'year': 'Year_Ceremony', 'status': 'Status', 'award': 'Golden_Globe_Award', 'movie': 'Movie_Name'}, inplace=True)
goldenglobe_data.head(5)
```

```python
# Remove awards that contain "Television" in the string
## We are only looking at the movie golden globe awards
goldenglobe_data = goldenglobe_data[~goldenglobe_data['Golden_Globe_Award'].str.contains("Television", na=False)]
goldenglobe_data.head(5)
```

```python
# Create DataFrame of Golden Globe winners.
goldenglobe_winners_df = goldenglobe_data[goldenglobe_data['Status'] == "Winner"][['Winner', 'Year_Ceremony', 'Status', 'Golden_Globe_Award', 'Movie_Name']]
goldenglobe_winners_df.head(5)
```

```python
# Check for duplicates
duplicate_rows = goldenglobe_winners_df[goldenglobe_winners_df.duplicated(subset=['Winner', 'Golden_Globe_Award'])]
# Display duplicate rows
duplicate_rows
```

```python
# Drop duplicate rows
gg_award1_df = goldenglobe_winners_df.drop_duplicates(subset=['Winner', 'Golden_Globe_Award'])

# Drop rows for a year 2024 (IMdB data is up to 2023)
gg_award_df = gg_award1_df[gg_award1_df['Year_Ceremony'] != 2024]

# Display DataFrame without duplicates
gg_award_df.head(5)
```

```python
# Check for unique awards
unique_awards = gg_award_df['Golden_Globe_Award'].unique()
print(unique_awards)
print(len(unique_awards))
```

```python
# Define IDs for each unique award
gg_award_ids = [
    'GG001', 'GG002', 'GG003', 'GG004', 'GG005', 'GG006', 'GG007', 'GG008', 'GG009', 'GG010','GG011', 'GG012', 'GG013', 'GG014', 'GG015', 'GG016', 'GG017', 'GG018', 'GG019', 'GG020', 'GG021', 'GG022', 'GG023',
    'GG024', 'GG025', 'GG026', 'GG027', 'GG028', 'GG029', 'GG030', 'GG031', 'GG032', 'GG033', 'GG034', 'GG035', 'GG036', 'GG037', 'GG038', 'GG039', 'GG040', 'GG041']

# Create a dictionary to map awards to their IDs
award_id_mapping = dict(zip(unique_awards, gg_award_ids))

# Add a new column with award IDs to the DataFrame
gg_award_df['GG_Award_ID'] = gg_award_df['Golden_Globe_Award'].map(award_id_mapping)

# Display the updated DataFrame
gg_award_df.head(5)
```

```python
# Copy values from 'Winners' column to 'Movie Name' column where 'Movie Name' is NaN
gg_award_df.loc[gg_award_df['Movie_Name'].isna(), 'Movie_Name'] = gg_award_df.loc[gg_award_df['Movie_Name'].isna(), 'Winner']

# Display the DataFrame with updated 'Movie Name' column
gg_award_df.head(5)
```

```python
# Function to rearrange movie name column
def rearrange_movie_name(movie_name):
    if isinstance(movie_name, str):  
        if movie_name.endswith(', The'):  
            movie_title = movie_name[:-5]  # Remove ', The' from the end of the string
            return 'The ' + movie_title
        elif movie_name.endswith(', A'):  
            movie_title = movie_name[:-3]  # Remove ', A' from the end of the string
            return 'A ' + movie_title
    return movie_name

# Rearrange the 'Movie Name' column
gg_award_df['Rearranged_Movie_Name'] = gg_award_df['Movie_Name'].apply(rearrange_movie_name)

# Display the DataFrame with rearranged 'Movie Name' column
gg_award_df.head(5)
```

```python
# Check the rows with null value. (decide to leave it because it's not significant)
rows_with_null = gg_award_df[gg_award_df.isnull().any(axis=1)]
rows_with_null.head(10)
```

```python
# Drop the column "Movie Name"
gg_award_df.drop(columns=["Movie_Name", "Status"], inplace=True)

# Rename the column "Rearranged Movie Name" to "Movie Name"
gg_award_df.rename(columns={"Rearranged_Movie_Name": "Movie_Name"}, inplace=True)

gg_award_df.head(10)
```

```python
# Add an ID for Golden Globe Winner
gg_award_df['Golden_Globe_Winner_ID'] = range(1, len(gg_award_df) + 1)
gg_award_df.head(10)
```

```python
# Remove the index column
df_without_index = gg_award_df.reset_index(drop=True)

# Reorder the columns 
columns_order = ['Golden_Globe_Winner_ID', 'Winner', 'Year_Ceremony', 'Movie_Name', 'Golden_Globe_Award', 'GG_Award_ID']
new_gg_award_df = df_without_index[columns_order]

# Display the updated DataFrame
new_gg_award_df.head(5)
```

```python
# Export goldenglobe_winners_df as CSV files.
new_gg_award_df.to_csv("CleanData/GoldenGlobe_Winners.csv", index=False)
```

```python
# Create a Golden Globe Awards DataFrame with the award_id array as the GG_award_id and awards list as the award category.
# Drop unwanted columns 
columns_to_drop = ['Golden_Globe_Winner_ID', 'Winner', 'Year_Ceremony', 'Movie_Name']
gg_award_id_df = new_gg_award_df.drop(columns=columns_to_drop)

# Drop duplicate rows
clean_gg_award_id_df = gg_award_id_df.drop_duplicates(subset=['Golden_Globe_Award', 'GG_Award_ID',])
clean_gg_award_id_df.head(5)
```

```python
# Remove the index column
df_without_index = clean_gg_award_id_df.reset_index(drop=True)

# Reorder the columns 
columns_order = ['GG_Award_ID', 'Golden_Globe_Award']
GG_df = df_without_index[columns_order]

# Display the updated DataFrame
GG_df.head(5)
```

```python
# Export Golden Globe Award list as CSV files.
GG_df.to_csv("CleanData/GG_AwardID.csv", index=False)
```
