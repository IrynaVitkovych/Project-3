```python
import pandas as pd
import numpy as np
```

```python
# Files to use
movies_data = pd.read_csv('Resources/movies.csv', encoding='utf-8')
oscar_winners = pd.read_csv('CleanData/Oscar_Winners.csv', encoding='utf-8')
goldenglobe_winners = pd.read_csv('CleanData/GoldenGlobe_Winners.csv', encoding='utf-8')

# Combine the data into a single DataFrame
combine_data_df = pd.merge(movies_data, oscar_winners, how='outer', on=["Movie_Name", "Movie_Name"])
complete_data_df = pd.merge(combine_data_df, goldenglobe_winners, how='outer', on=["Movie_Name", "Movie_Name"])

complete_data_df.head(10)
```

```python
# Columns name info
complete_data_df.columns
```

```python
# Drop unwanted columns 
columns_to_drop = ['Unnamed: 0', 'Director', 'Oscar_Winner_ID', 'Winner_x', 'Award', 'Year_Film', 'Year_Ceremony_x', 'Golden_Globe_Winner_ID', 'Winner_y', 'Year_Ceremony_y','Golden_Globe_Award']
complete_movies_df = complete_data_df.drop(columns=columns_to_drop)

# Convert columns to integers
columns_to_convert = ['Votes', 'Meta_score', 'Year', 'Movie_ID', 'Director_ID']
complete_movies_df[columns_to_convert] = complete_movies_df[columns_to_convert].fillna(0).astype(int)

complete_movies_df.head(10)
```

```python
# Drop duplicate rows
movies_df = complete_movies_df.drop_duplicates(subset=['Movie_Name', 'Oscar_Award_ID', 'GG_Award_ID'])

movies_df.head(10)
```

```python
unique_movies = movies_df['Movie_Name'].unique()
print(unique_movies)
print(len(unique_movies))
```

```python
# Export list of Oscar Winners as CSV files.
movies_df.to_csv("CleanData/Movies.csv", index=False)
```
