```python
import pandas as pd
import numpy as np
```

```python
# Files to use
movies_data = pd.read_csv('Resources/Movies.csv', encoding='utf-8')
oscar_winners = pd.read_csv('CleanData/Oscar_Winners.csv', encoding='utf-8')
goldenglobe_winners = pd.read_csv('CleanData/Golden_Globe_Winners.csv', encoding='utf-8')

# Combine the data into a single DataFrame
combine_data_df = pd.merge(movies_data, oscar_winners, how='outer', on=["Movie Name", "Movie Name"])
complete_data_df = pd.merge(combine_data_df, goldenglobe_winners, how='outer', on=["Movie Name", "Movie Name"])

complete_data_df.head(10)
```

```python
# Columns name info
complete_data_df.columns
```

```python
complete_data_df.info()
```

```python
# Drop unwanted columns 
columns_to_drop = ['Unnamed: 0', 'Director', 'Oscar Winner ID', 'Winner_x', 'Award', 'Year Film', 'Year Ceremony_x', 'Golden Globe Winner ID', 'Winner_y', 'Year Ceremony_y','Golden Globe Award']
complete_movies_df = complete_data_df.drop(columns=columns_to_drop)

# Rename column Year_x
complete_movies_df.rename(columns={'Year_x': 'Year'}, inplace=True)

# Convert columns to integers
columns_to_convert = ['Votes', 'Meta Score', 'Year', 'Movie ID', 'Director ID']
complete_movies_df[columns_to_convert] = complete_movies_df[columns_to_convert].fillna(0).astype(int)

complete_movies_df.head(10)
```

```python
# Drop duplicate rows
movies_df = complete_movies_df.drop_duplicates(subset=['Movie Name', 'Oscar Award ID', 'GG Award ID'])

movies_df.head(10)
```

```python
unique_movies = movies_df['Movie Name'].unique()
print(unique_movies)
print(len(unique_movies))
```

```python
# Export list of Oscar Winners as CSV files.
movies_df.to_csv("CleanData/MoviesWithAwards.csv", index=False)
```
